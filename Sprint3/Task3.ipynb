Создать базовый класс Event в C#, который будет представлять организованные
мероприятия. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Event:

• Атрибуты: ID мероприятия (EventId), Название мероприятия (EventName),
Дата проведения (Date).

• Методы:

    o RegisterParticipant(): метод для регистрации участника.
    o NotifyParticipants(): метод для уведомления участников о мероприятии.
    o GetEventDetails(): метод для получения деталей мероприятия.

Требования к производным классам:
1. Конференция (Conference): Должна содержать дополнительные атрибуты,
такие как Количество спикеров (SpeakersCount).
Метод NotifyParticipants() должен быть переопределен для отправки
уведомлений с программой конференции.
2. Фестиваль (Festival): Должна содержать дополнительные атрибуты, такие как
количество музыкальных групп (BandsCount).
Метод RegisterParticipant() должен быть переопределен для регистрации
участников с выбором интересующих музыкальных жанров.
3. Семинар (Seminar) (если требуется третий класс): Должна содержать
дополнительные атрибуты, такие как Тематические области (Topics).
Метод GetEventDetails() должен быть переопределен для отображения
тематических областей семинара вместе с другими деталями мероприятия.

```
public interface IEvent 
{
    int EventId { get; set; }
    string EventName { get; set; }
    DateTime DateTime { get; set; }

    void NotifyParticipants();
    void RegisterParticipant(string participant);
    void GetEventDetails();
}

public interface IContest
{
    void Addition(string Privet);
}

public interface IConference : IEvent { int SpeakersCount { get; set; } }
public interface IFestival : IEvent { int BandsCount { get; set; } }
public interface ISeminar : IEvent { string Topics { get; set; } }

public class Event : IEvent
{
    public int EventId { get; set; }
    public string EventName { get; set; }
    public DateTime DateTime { get; set; }
    public List<string> Participants { get; set; }

    public Event(int eventId, string eventName, DateTime dateTime)
    { EventId = eventId; EventName = eventName; DateTime = dateTime; Participants = new List<string>(); }

    /// o RegisterParticipant(): метод для регистрации участника.
    /// o NotifyParticipants(): метод для уведомления участников о мероприятии.
    /// o GetEventDetails(): метод для получения деталей мероприятия.
    
    public virtual void RegisterParticipant(string participant)
    {
        Participants.Add(participant);
        Console.WriteLine($"{participant} зарегистрирован на мероприятие «{EventName}».");
    }

    public virtual void NotifyParticipants() 
    {
        foreach (var participant in Participants)
        {
            Console.WriteLine($"Уведомление для {participant}: «{EventName}» пройдет {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()}.");
        }
    }
    
    public virtual void GetEventDetails() 
    {
        Console.WriteLine($"Айди: {EventId} | Мероприятие: «{EventName}» | Дата проведения: {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()}.");
    }

    public virtual void Addition(int Num)
    {
        Num++;
        Console.WriteLine(Num);
    }

    public virtual void Addition(double CharNum)
    {
        CharNum++; CharNum = Math.Round(CharNum);
        Console.WriteLine(CharNum);
    }
}

///Sprint3
public class EventCollection<T> where T : Event
{
    private List<T> events = new List<T>();

    public void Add(T event1)
    {
        events.Add(event1);
    }
    
    public void Remove(T event2)
    {
        events.Remove(event2);
    }
    
    public void DisplayEvents()
    {
        foreach (var event0 in events)
        {
            event0.NotifyParticipants();
            event0.GetEventDetails();
        }
    }

    public static EventCollection<T> operator +(EventCollection<T> collection1, EventCollection<T> collection2)
    {
        EventCollection<T> comboP = new EventCollection<T>();

        foreach (var event1 in collection1.events)
        {
            comboP.Add(event1);
        }

        foreach (var event2 in collection2.events)
        {
            comboP.Add(event2);
        }

        return comboP;
    }
}

/// Sprint2
public class Contest : Event, IContest
{
    
    public Contest(int eventId, string eventName, DateTime dateTime, string privet, int num, double charNum)
     : base (eventId, eventName, dateTime) {}

    public override void Addition(int num)
    {
        num++;
        Console.WriteLine($"Зарегистрирован. Число зарегистрированных: {num}");
    }

    public void Addition(string privet)
    {
        Console.WriteLine(privet);
    }

    public override void Addition(double charNum)
    {
        charNum++; charNum = Math.Round(charNum);
        Console.WriteLine($"Зарегистрирован. Число зарегистрированных: {charNum}");
    }
}

public class Conference : Event, IConference 
{
    public int SpeakersCount { get; set; }
    
    public Conference(int eventId, string eventName, DateTime dateTime, int speakersCount) 
     : base (eventId, eventName, dateTime) { SpeakersCount = speakersCount; }
    
    /// o NotifyParticipants(): для отправки уведомлений с программой конференции.

    public override void NotifyParticipants()
    {
        foreach (var participant in Participants)
        {
            Console.WriteLine($"Уведомление для {participant}: Конференция «{EventName}» пройдет {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()} с участием {SpeakersCount} спикеров. С программой конферении можете ознакомиться ниже");
        }
    }

    public void AddParticipants(Festival value)
    {
        /// + value.Participants();
        Participants.AddRange(value.Participants);
    }
}

public class Festival : Event, IFestival 
{
    public int BandsCount { get; set; }
    public static List<string> Genres { get; private set; } = new List<string>{ "Rock", "Pop", "Jazz", "Electronic" };

    public Festival(int eventId, string eventName, DateTime dateTime, int bandsCount)
     : base(eventId, eventName, dateTime) { BandsCount = bandsCount; }

    /// o RegisterParticipants(): для регистрации участников с выбором интересующих музыкальных жанров.

    public override void RegisterParticipant(string participant)
    {
        Console.WriteLine($"{participant}, выберите интересующий музыкальный жанр:");
        Participants.Add(participant);
        Console.WriteLine(String.Join("\t", Genres));
        
        ///int choice1 = int.Parse("4");
        string chosenGenre = Genres[int.Parse("4") - 1];
        Console.WriteLine("Выбор сделан.");
        Console.WriteLine($"{participant} зарегистрирован на фестиваль «{EventName}», жанр которого {chosenGenre}, где количество групп за время проведения меропрития составит {BandsCount}");
    }
}

public class Seminar : Event, ISeminar
{
    public string Topics { get; set; }

    public Seminar(int eventId, string eventName, DateTime dateTime, string topics)
     : base (eventId, eventName, dateTime) { Topics = topics; }

    /// o GetEventDetails(): для отображения тематических областей семинара вместе с другими деталями мероприятия.

    public override void GetEventDetails() 
    {
        Console.WriteLine($"Айди: {EventId} | Название мероприятия: «{EventName}» | Тематическая область: {Topics} | Дата проведения: {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()}");
    }
}

public class SportingEvent : Event
{
    public int CommentatorCount { get; set; }
    public string Commentator { get; set; }

    public SportingEvent(int eventId, string eventName, DateTime dateTime, int commentatorCount, string commentator)
     : base (eventId, eventName, dateTime) { CommentatorCount = commentatorCount; Commentator = commentator; }

    /// o RegisterParticipant(): метод для регистрации участника.
    /// o NotifyParticipants(): метод для уведомления участников о мероприятии.
    /// o GetEventDetails(): метод для получения деталей мероприятия.

    public override void NotifyParticipants()
    {
        foreach (var participant in Participants)
        {
            Console.WriteLine($"Уведомление для {participant}: Спортивное мероприятие «{EventName}» пройдет {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()} с участием {CommentatorCount} комментаторов.");
        }
    }
    
    public override void RegisterParticipant(string participant)
    {
        Participants.Add(participant);
        Console.WriteLine($"{participant} зарегистрирован на спортивное меропрятие «{EventName}», где количество комментаторов за время проведения меропрития составит {CommentatorCount}");
    }
    
    public override void GetEventDetails()
    {
        string [] commentator = Commentator.Split(",");
        foreach (var comment in commentator)
        {
            Console.WriteLine($"Айди: {EventId} | Название мероприятия: «{EventName}» | Дата проведения: {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()} | Комментаторы: {comment}");
        }
       
    }
}

public class Exhibition : Event
{
    public string ExhibitCount { get; set; }

    public Exhibition(int eventId, string eventName, DateTime dateTime, string exhibitCount)
     : base (eventId, eventName, dateTime) { ExhibitCount = exhibitCount; }

    public override void GetEventDetails()
    {
        Console.WriteLine($"Айди: {EventId} | Название мероприятия: «{EventName}» | Дата проведения: {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()} | Количество экспонатов: {ExhibitCount}");
    }
}

public class TypesExhibition : Exhibition
{
    public string TypExhibition {get; set;}

    public TypesExhibition(int eventId, string eventName, DateTime dateTime, string exhibitCount, string typExhibition)
     : base (eventId, eventName, dateTime, exhibitCount) { TypExhibition = typExhibition; }

    public override void GetEventDetails()
    {
        ///base.ExhibitCount = "15";
        Console.WriteLine($"Айди: {EventId} | Название мероприятия: «{EventName}» | Тип мероприятия: {TypExhibition} | Дата проведения: {DateTime.ToShortDateString()} {DateTime.ToShortTimeString()} | Количество экспонатов: {ExhibitCount}");
    }
}

var eevent = new Event(4, "Conference", new DateTime(2024, 9, 15, 9, 30, 0));
var conference = new Conference(1, "Tech Conference", new DateTime(2024, 9, 14, 15, 0, 0), 5);
var festival = new Festival(2, "Music Festival", new DateTime(2024, 9, 10, 18, 25, 0), 10);
var seminar = new Seminar(3, "AI Seminar", new DateTime(2024, 10, 5, 10, 0, 0), "AI Ethics");
var sportingEvent = new SportingEvent(4, "Ставки на спорт", new DateTime(2024, 06, 10, 22, 0, 0), 2, "Gimmi, Rimmi");
var exhibition = new Exhibition(5, "Streets Moscow", new DateTime(2024, 10, 5, 12, 0, 0), "15");
var typesExhibition = new TypesExhibition(6, "Village Paris", new DateTime(2024, 10, 5, 18, 0, 0), "15", "Тематический");
var contest = new Contest(7, "Seminar", new DateTime(2024, 9, 15, 9, 30, 0), "Hello", 15, 18.5);

EventCollection<Event> event1 = new EventCollection<Event>();
EventCollection<Event> event2 = new EventCollection<Event>();
var eventt1 = new Event(10, "Machine", new DateTime(2024, 9, 15, 9, 30, 0));
var eventt2 = new Event(11, "Shark", new DateTime(2024, 12, 15, 9, 30, 0));
event1.Add(eventt1);
event2.Add(eventt2);
EventCollection<Event> event3 = event1 + event2;

EventCollection<Event> seminars = new EventCollection<Event>();
EventCollection<Event> sportingEvent1 = new EventCollection<Event>();

Console.WriteLine("_______________________________________________________________________________________________________________________________");

seminars.Add(seminar);
sportingEvent1.Add(sportingEvent);
seminars.DisplayEvents();
sportingEvent1.DisplayEvents();

Console.WriteLine("_______________________________________________________________________________________________________________________________");

contest.Addition(15);
contest.Addition("Hello");
contest.Addition(18.5);

Console.WriteLine("_______________________________________________________________________________________________________________________________");
```
